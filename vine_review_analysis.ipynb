```python
import pandas as pd
```

```python
vine_df = pd.read_csv("vine_table.csv", low_memory=False)
vine_df.head()
```

```python
vine_df.dtypes
```

```python
vote20_df = vine_df.loc[vine_df["total_votes"] >= 20]
vote20_df
```

```python
df = vote20_df.loc[vote20_df["helpful_votes"]/vote20_df["total_votes"] >= 0.5]
df.head()
```

```python
yes_df = df.loc[df["vine"]=="Y"]
yes_df.head()
```

```python
no_df = df.loc[df["vine"]=="N"]
no_df.head()
```

```python
rate_5_star_num = yes_df.loc[(yes_df['star_rating']=='5')]["star_rating"].count()

total_rating = yes_df["star_rating"].count()

percentage_5_star_vine =rate_5_star_num / total_rating * 100

print(f"5 Star Rating: {rate_5_star_num}")
print(f"Total Rating: {total_rating}")
print(f"Percentage of 5 star: {percentage_5_star_vine}")
```

```python
five_star_number = no_df.loc[(no_df['star_rating']== '5')]["star_rating"].count()
 
unpaid_number = no_df["star_rating"].count()

percentage_five_star_non_vine = five_star_number / unpaid_number * 100

print(f"5 Star Rating: {five_star_number}")
print(f"Total Rating: {unpaid_number}")
print(f"Percentage of 5 star: {percentage_five_star_non_vine}")
```
